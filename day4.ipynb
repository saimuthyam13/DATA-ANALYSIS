```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split as ttp
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
import re
import string
import nltk
```

```python
data_alive=pd.read_csv("/content/alive.csv")
data_notalive=pd.read_csv("/content/notalive.csv")
data_alive.head(5)
data_notalive.head(5)
data_alive.shape,data_notalive.shape
data_alive["class"]=1
data_notalive["class"]=0
data_alive_test=data_alive.tail(10)
for i in range(24,14,-1):
  data_alive.drop([i],axis=0,inplace=True)
data_notalive_test=data_notalive.tail(10)
for i in range(24,14,-1):
  data_notalive.drop([i],axis=0,inplace=True)
data_merge=pd.concat([data_notalive,data_alive],axis=0)
data_merge.to_csv("manual_training.csv")
dmt=pd.concat([data_notalive_test,data_alive_test],axis=0)
dmt.to_csv("manual_testing.csv")
```

```python
#linear Regression
data_training=pd.read_csv('/content/manual_training.csv')
data_testing=pd.read_csv('/content/manual_testing.csv')
x1=data_training[["PassengerId"]]
y1=data_training["class"]
x1.shape
y1.shape
x_test=data_testing["PassengerId"]
y_test=data_testing["class"]
model=LinearRegression()
model.fit(x1,y1)
x_test = x_test.values.reshape(-1, 1)
prediction=model.predict(x_test)
plt.scatter(x_test,prediction)
plt.title("Titanic Survival Rate")
plt.xlabel("PassengerId")
plt.ylabel("Survived")
plt.show()
```

```python
#l0gistic Regression
data_training=pd.read_csv('/content/manual_training.csv')
data_testing=pd.read_csv('/content/manual_testing.csv')
x1=data_training[["PassengerId"]]
y1=data_training["class"]
x1.shape
y1.shape
x_test=data_testing["PassengerId"]
y_test=data_testing["class"]
model=LogisticRegression()
model.fit(x1,y1)
x_test = x_test.values.reshape(-1, 1)
prediction=model.predict(x_test)
plt.scatter(x_test,prediction)
print(prediction)
plt.title("Titanic Survival Rate")
plt.xlabel("PassengerId")
plt.ylabel("Survived")
plt.show()
```
